```python
import numpy as np
import pandas as pd
```

```python
from scipy import stats
```

```python
df=pd.read_csv('grades-Copy1.csv')
df.head()
```

```python
print("Há {} linhas e {} colunas no dataset".format(df.shape[0], df.shape[1]))
```

```python
primeiros_finalistas=df[pd.to_datetime(df['assignment1_submission']) < '2016']
```

```python
primeiros_finalistas
```

```python
ultimos_finalistas=df[~df.index.isin(primeiros_finalistas.index)] #pego o inverso dos primeiros finalistas 
ultimos_finalistas.head()
```

```python
#qual a media das notas em cada grupo
print(primeiros_finalistas['assignment1_grade'].mean())
print(ultimos_finalistas['assignment1_grade'].mean())
```

as medias estão muito parecidas, é aqui que entra o teste de hipoteses t-students. Ele possibilita formar a hipotese alternativa (não há diferença) e a hipotese nula (as medias são as mesmas) e então testar essa hipótese nula.

Para o teste de hipotese se escolhe um nível de significância como um limite para quanto de uma chance estamos dispostos a aceitar, conhecido como ALFA

```python
from scipy.stats import ttest_ind
```

```python
ttest_ind(primeiros_finalistas['assignment1_grade'], ultimos_finalistas['assignment1_grade'])
```

com essa função, vemos que a probabilidade da hipotese nula é de 18%. Isso esta acima de um alfa aceitavel (definido arbitrariamente) de 0,05. Assim, não rejeito a hipotese nula, porque nao ha certeza suficiente para afirmar que as duas amostras são as mesmas.

```python
#olhando para os valores de p, para cada linha em dois dataframes distintos, elas são as mesmas colunas?
#Simulo então uma comparação aleatória entre dois dataframes, considerando o alfa de 10%
#Assim, sinalizo quando o valor-p não for inferior a 10%
df1=pd.DataFrame([np.random.random(100) for x in range(100)])
df1.head()
```

```python
df2=pd.DataFrame([np.random.random(100) for x in range(100)])
```

```python
def test_columns(alpha=0.1):
    num_diff=0
    for col in df1.columns:
        teststat,pval=ttest_ind(df1[col],df2[col])
        if pval<=alpha:
            print("A coluna {} é estatisticamente significativamente diferente do alpha={}, valor p={}".format(col,alpha,pval))
            num_diff=num_diff+1
    print("O número total diferente foi {}, do qual é {}%".format(num_diff,float(num_diff)/len(df1.columns)*100))
```

```python
test_columns()
```
